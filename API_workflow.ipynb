```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.RandomState(529)
```

```python
x = 90 * rng.rand(50)
print(x)
```

```python
y = 18 * x - 1 + rng.randn(50)
print(y)
```

```python
plt.scatter(x,y);
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model = LinearRegression(fit_intercept=True)
```

```python
model
```

```python
X = x[:, np.newaxis]
X.shape
```

```python
model.fit(X,y)
```

```python
xnew = np.linspace(-9, 99)
Xnew = xnew[:, np.newaxis]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
ynew = model.predict(Xnew)
ynew.shape
```

```python
model.coef_
```

```python
model.intercept_
```

```python
xfit = np.linspace(-1,11)
xfit.ndim
```

```python
Xfit=xfit[:,np.newaxis]
Xfit.ndim
```

```python
yfit = model.predict(Xfit)
yfit.ndim
```

```python
plt.scatter(x, y, label="Data points")
plt.plot(xnew, ynew, color="red", label="Linear fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Fit")
plt.grid(True)
plt.show()
```
